### Importing the Dataset


```python
import pandas as pd
maindf = pd.read_csv('content/BTC-USD.csv')
```

### Data Preprocessing:
Normalizing the Close price data using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler

# Select 'Close' price
data = maindf.filter(['Close'])

# Convert the dataframe to a numpy array
dataset = data.values

# Scale the data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
```

### Training Data Creation
Createing a dataset where X_train contains 'n' past values and y_train contains the 'n+1' value.

```python
import numpy as np

# Create the training data set
train_data = scaled_data[0:int(len(dataset)*0.8), :]

# Split the data into x_train and y_train data sets
x_train, y_train = [], []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])

# Convert x_train and y_train to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data into the shape accepted by the LSTM
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

### Model Creation
Creating the LSTM Model using keras

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape= (x_train.shape[1], 1)))
model.add(LSTM(50, return_sequences= False))
model.add(Dense(25))
model.add(Dense(1))
```

### Model Compliation and Training

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
from keras.callbacks import ModelCheckpoint

# Define the checkpoint
filepath = "best_model.h5"
checkpoint = ModelCheckpoint(filepath, monitor='val_mean_absolute_error', verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint]

# Fit the model
history = model.fit(x_train, y_train, batch_size=1, epochs=20, validation_split=0.2, callbacks=callbacks_list)
```

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()

# Plot training & validation mean absolute error values
plt.figure(figsize=(12, 6))
plt.plot(history.history['mean_absolute_error'])
plt.plot(history.history['val_mean_absolute_error'])
plt.title('Mean Absolute Error')
plt.ylabel('Mean Absolute Error')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```

### Testing Data Creation: 
Creating a dataset where X_test contains 'n' past values from the training data and the rest from the testing data.

```python
# Create the testing data set
test_data = scaled_data[int(len(dataset)*0.8) - 60:, :]

# Create the data sets x_test and y_test
x_test, y_test = [], dataset[int(len(dataset)*0.8):, :]

for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])

# Convert x_test to a numpy array 
x_test = np.array(x_test)

# Reshape the data into the shape accepted by the LSTM
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

### Prediction and Visualization

```python
import matplotlib.pyplot as plt

# Get the models predicted price values
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Plot the data
plt.figure(figsize=(10,6))
plt.plot(data[int(len(dataset)*0.8):].index, y_test, color='blue', label='Actual Bitcoin Price')
plt.plot(data[int(len(dataset)*0.8):].index, predictions, color='red', label='Predicted Bitcoin Price')
plt.title('Bitcoin Price Prediction')
plt.xlabel('Date')
plt.ylabel('Bitcoin Price')
plt.legend()
plt.show()
```

### Comparison of the Best model with others

```python
from keras.models import load_model

# Load the best model
best_model = load_model('best_model.h5')

# Get the best model's predicted price values
best_predictions = best_model.predict(x_test)
best_predictions = scaler.inverse_transform(best_predictions)

# Plot the data
plt.figure(figsize=(10,6))
plt.plot(data[int(len(dataset)*0.8):].index, y_test, color='blue', label='Actual Bitcoin Price')
plt.plot(data[int(len(dataset)*0.8):].index, predictions, color='red', label='Predicted Bitcoin Price by Current Model')
plt.plot(data[int(len(dataset)*0.8):].index, best_predictions, color='green', label='Predicted Bitcoin Price by Best Model')
plt.title('Bitcoin Price Prediction')
plt.xlabel('Date')
plt.ylabel('Bitcoin Price')
plt.legend()
plt.show()
```
